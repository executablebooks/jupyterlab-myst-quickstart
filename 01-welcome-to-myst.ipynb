MyST (<https://mystmd.org>) is an ecosystem of open-source, community-driven tools designed for scientific communication, including a powerful authoring framework that supports blogs, online books, scientific papers, reports and journals articles!

🎉 MyST now works in JupyterLab 4.0, see [#142](https://github.com/executablebooks/jupyterlab-myst/pull/142)! 🎉

## Welcome to using MyST in Jupyter

:::{important} Installing MyST Markdown
These notebooks should be rendered with `jupyterlab_myst` ([JupyterLab extension](https://github.com/executablebooks/jupyterlab-myst)) or `myst` ([command line tools](https://mystmd.org/guide)).

```bash
# For working in JupyterLab
pip install jupyterlab_myst
# For working on the command-line
npm install -g mystmd
```
:::

:::{tip} Verifying the `jupyterlab_myst` Installation
:class: dropdown
To verify that the extension is registered with Jupyter, use:

```text
jupyter labextension list
```

You should see the following text in the output:

```text
jupyterlab-myst v2.x.x enabled OK
```

In Jupyter 4.0 you can also see the MyST plugin in the Jupyter Extension Manager inside of JupyterLab.

:::{figure} attachment:5d201865-319b-49cf-affa-07789d045463.png
:width: 50%
In the Jupyter Extension Manager you will see that the MyST extension is installed and enabled.
:::
